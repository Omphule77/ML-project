```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv')
dataset.head()
```

```python
dataset.info()
```

```python
## Data cleaning
## Finding the missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
## The dataset converted into two set based on the region from 123th of index.
## we can make a new column based on region
## 1. Bejaia Region Dataset
##2. Sidi-Bel Abbes Region Dataset
## Add new column with region
```

```python
dataset.loc[:123,'Region']=0
dataset.loc[123:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
## removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## we need to drom 122th row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
## fix the columns spaces
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
## chenges the required column as int datatype
df.info()
```

```python
df.head()
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]=df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
## lets save the this cleaned CSV file.
df.to_csv('Algerian_forest_fires_cleaned_dataset',index=False) ## index=false means starting numberical index also save so for avoid this use false
```
